```python
# COMPLETE CODE
# First, let's add the missing portfolio_performance function and data loading code
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import minimize
import yfinance as yf

# Step 1: Define tickers and fetch data
tickers = ['AAPL', 'MSFT', 'GOOG', 'AMZN', 'META']  # Example tickers
start_date = '2018-01-01'
end_date = '2023-01-01'

# Download data - using 'Close' instead of 'Adj Close'
data = yf.download(tickers, start=start_date, end=end_date)
# Check if 'Adj Close' exists, if not use 'Close'
if 'Adj Close' in data.columns:
    prices = data['Adj Close']
else:
    prices = data['Close']  # Use 'Close' if 'Adj Close' is not available

# Step 2: Calculate returns
returns = prices.pct_change().dropna()
mean_returns = returns.mean()
cov_matrix = returns.cov()

# Step 3: Portfolio Performance function
def portfolio_performance(weights, mean_returns, cov_matrix):
    returns = np.sum(mean_returns * weights) * 252  # Annualized return
    std = np.sqrt(np.dot(weights.T, np.dot(cov_matrix, weights))) * np.sqrt(252)  # Annualized volatility
    return returns, std

# Step 4: Negative Sharpe Ratio (to minimize)
def neg_sharpe_ratio(weights, mean_returns, cov_matrix, risk_free_rate=0.0):
    ret, std = portfolio_performance(weights, mean_returns, cov_matrix)
    return -((ret - risk_free_rate) / std)

# Step 5: Constraints and bounds
def check_sum(weights):
    return np.sum(weights) - 1

bounds = tuple((0, 1) for _ in range(len(tickers)))
initial_guess = len(tickers) * [1. / len(tickers)]
constraints = ({'type': 'eq', 'fun': check_sum})

# Step 6: Optimization
optimal = minimize(neg_sharpe_ratio, initial_guess, args=(mean_returns, cov_matrix),
                   method='SLSQP', bounds=bounds, constraints=constraints)

optimal_weights = optimal.x
opt_return, opt_std = portfolio_performance(optimal_weights, mean_returns, cov_matrix)
opt_sharpe = (opt_return) / opt_std

# Step 7: Simulate random portfolios
num_portfolios = 10000
results = np.zeros((3, num_portfolios))
weight_array = []

for i in range(num_portfolios):
    weights = np.random.random(len(tickers))
    weights /= np.sum(weights)
    weight_array.append(weights)
    ret, std = portfolio_performance(weights, mean_returns, cov_matrix)
    results[0,i] = std
    results[1,i] = ret
    results[2,i] = (ret) / std

# Step 8: Plot Efficient Frontier
plt.figure(figsize=(12, 6))
plt.scatter(results[0,:], results[1,:], c=results[2,:], cmap='viridis', marker='o', s=10, alpha=0.3)
plt.colorbar(label='Sharpe Ratio')
plt.scatter(opt_std, opt_return, c='red', marker='*', s=300, label='Optimal Portfolio')
plt.title('Efficient Frontier with Optimal Portfolio')
plt.xlabel('Annualized Volatility')
plt.ylabel('Annualized Return')
plt.legend(labelspacing=0.8)
plt.grid(True)
plt.show()

# Step 9: Print results
print("Optimal Portfolio Allocation:")
for i, ticker in enumerate(tickers):
    print(f"{ticker}: {optimal_weights[i]:.2%}")
print(f"Expected Annual Return: {opt_return:.2%}")
print(f"Annual Volatility (Risk): {opt_std:.2%}")
print(f"Sharpe Ratio: {opt_sharpe:.2f}") 
```
